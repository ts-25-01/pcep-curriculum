#### üìù Spezielle Parameter

Die `print()`-Funktion akzeptiert mehrere spezielle Parameter, die ihr Verhalten anpassen:

##### `sep` Parameter

Der `sep`-Parameter (separator) definiert, wie die Argumente getrennt werden. Standardm√§√üig ist dies ein Leerzeichen:


````python
print("Hello", "World")           # Gibt "Hello World" aus
print("Hello", "World", sep="")   # Gibt "HelloWorld" aus
print("Hello", "World", sep="-")  # Gibt "Hello-World" aus
print("Hello", "World", sep="\n") # Gibt "Hello" und "World" in separaten Zeilen aus
````

##### `end` Parameter

Der `end`-Parameter bestimmt, was am Ende der Ausgabe ausgegeben wird. Standardm√§√üig ist dies ein Zeilenumbruch (`\n`):


````python
print("Hello", end="")     # Kein Zeilenumbruch am Ende
print("World")             # Ausgabe: "HelloWorld"

print("Hello", end=" ")    # Leerzeichen am Ende statt Zeilenumbruch
print("World")             # Ausgabe: "Hello World"

print("Hello", end="!\n")  # Ausrufezeichen und Zeilenumbruch am Ende
print("World")             # Ausgabe: "Hello!" und dann "World" in der n√§chsten Zeile
````

##### `file` Parameter

Der `file`-Parameter bestimmt, wohin die Ausgabe gesendet wird. Standardm√§√üig ist dies `sys.stdout` (die Standardausgabe):


````python
# In eine Datei schreiben
with open("output.txt", "w") as f:
    print("Diese Zeile wird in eine Datei geschrieben", file=f)
````

##### `flush` Parameter

Der `flush`-Parameter steuert, ob der Ausgabepuffer sofort geleert wird. Standardm√§√üig ist dieser `False`:


````python
import time
print("L√§dt", end="", flush=True)
for i in range(3):
    time.sleep(1)  # Eine Sekunde warten
    print(".", end="", flush=True)
print(" Fertig!")
````

#### üìù 2. %-Formatierung (alte Methode)

Die %-Formatierung war lange Zeit die Standardmethode in Python:

````python
name = "Bob"
alter = 25
begruessung = "Hallo, %s! Du bist %d Jahre alt." % (name, alter)
print(begruessung)
````

Platzhalter in der %-Formatierung:

- `%s` f√ºr Strings
- `%d` f√ºr Ganzzahlen
- `%f` f√ºr Gleitkommazahlen
- `%.2f` f√ºr Gleitkommazahlen mit 2 Dezimalstellen
- `%x` f√ºr hexadezimale Ganzzahlen
- `%o` f√ºr oktale Ganzzahlen
- `%e` f√ºr wissenschaftliche Notation
- `%g` f√ºr allgemeine Formatierung
- `%c` f√ºr ein einzelnes Zeichen
- `%u` f√ºr unsigned integers (nicht in Python 3)
- `%p` f√ºr Zeiger (nicht in Python 3)
- `%n` f√ºr die Anzahl der bereits ausgegebenen Zeichen (nicht in Python 3)
- `%b` f√ºr bin√§re Ganzzahlen (nicht in Python 3)
- `%a` f√ºr hexadezimale Gleitkommazahlen (nicht in Python 3)
- `%r` f√ºr die `repr()`-Darstellung eines Objekts

````python
# Formatierung mit Pr√§zision
preis = 19.99
menge = 3
gesamt = preis * menge
print("Preis: %.2f ‚Ç¨ √ó %d = %.2f ‚Ç¨" % (preis, menge, gesamt))
````

#### üìù 3. str.format() Methode

Die `str.format()`-Methode bietet mehr Flexibilit√§t:


````python
name = "Charlie"
alter = 35
begruessung = "Hallo, {}! Du bist {} Jahre alt.".format(name, alter)
print(begruessung)
````

Die `format()`-Methode erlaubt auch die Verwendung von Indexen und benannten Parametern:

````python
# Mit Indexen
text = "{0} liebt {1}, aber {1} liebt {2}".format("Romeo", "Julia", "Paris")
print(text)  # "Romeo liebt Julia, aber Julia liebt Paris"

# Mit benannten Parametern
text = "{name} ist {alter} Jahre alt und lebt in {stadt}".format(
    name="David", alter=28, stadt="Berlin")
print(text)  # "David ist 28 Jahre alt und lebt in Berlin"
````

Formatierungsoptionen:


````python
# Zahlenformatierung
pi = 3.14159
print("Pi mit 2 Nachkommastellen: {:.2f}".format(pi))  # 3.14
print("Pi als Prozent: {:.2%}".format(pi/100))         # 3.14%

# Ausrichtung und F√ºllzeichen
for i in range(1, 11):
    print("Nummer {0:2d}: {0:3d} hoch 2 ist {1:4d}".format(i, i**2))
````

```python
if bedingung:
```

Die Bedingung wird ausgewertet und wenn sie zu `True` evaluiert, wird der einger√ºckte Codeblock ausgef√ºhrt.


````python
alter = 18
if alter >= 18:
    print("Du bist vollj√§hrig.")
````

Mit der `if-else`-Anweisung k√∂nnen wir einen alternativen Codeblock ausf√ºhren, wenn die Bedingung nicht erf√ºllt ist:

````python
if alter >= 18:
    print("Du bist vollj√§hrig.")
    print("Du darfst w√§hlen.")
print("Diese Zeile wird immer ausgegeben.")  # Nicht Teil des if-Blocks
````

### Verschachtelte Bedingungen

Verschachtelte Bedingungen sind bedingte Anweisungen innerhalb anderer bedingter Anweisungen. Sie erm√∂glichen es, komplexere Entscheidungsstrukturen zu erstellen.

#### üìù Syntax f√ºr verschachtelte Bedingungen
```python
if √§u√üere_bedingung:
    # Code f√ºr √§u√üere Bedingung
    if innere_bedingung:
        # Code f√ºr innere Bedingung
    else:
        # Code, wenn innere Bedingung nicht erf√ºllt ist
else:
    # Code, wenn √§u√üere Bedingung nicht erf√ºllt ist
```

#### üìù Beispiele f√ºr verschachtelte Bedingungen

**Beispiel 1: Zahlenklassifikation**


````python
zahl = int(input("Gib eine ganze Zahl ein: "))

if zahl > 0:
    print("Die Zahl ist positiv.")
    if zahl % 2 == 0:
        print("Die Zahl ist gerade.")
    else:
        print("Die Zahl ist ungerade.")
else:
    if zahl < 0:
        print("Die Zahl ist negativ.")
        if zahl % 2 == 0:
            print("Die Zahl ist gerade.")
        else:
            print("Die Zahl ist ungerade.")
    else:
        print("Die Zahl ist Null.")
````

**Beispiel 2: Benutzerauthentifizierung**


````python
benutzername = input("Benutzername: ")
passwort = input("Passwort: ")

if benutzername == "admin":
    if passwort == "12345":
        print("Administratorzugriff gew√§hrt.")
    else:
        print("Falsches Passwort f√ºr Administrator.")
else:
    if benutzername == "gast":
        if passwort == "gast":
            print("Gastzugriff gew√§hrt.")
        else:
            print("Falsches Passwort f√ºr Gast.")
    else:
        print("Unbekannter Benutzer.")
````

#### üìù Tipps zur Verwendung verschachtelter Bedingungen

1. **Lesbarkeit**: Verschachtelte Bedingungen k√∂nnen schnell komplex und schwer zu lesen werden. Verwenden Sie sie mit Bedacht.
2. **Einr√ºckung**: Achten Sie besonders auf die korrekte Einr√ºckung bei verschachtelten Bedingungen, da dies in Python entscheidend f√ºr die Struktur ist.
3. **Alternativen**: Oft k√∂nnen verschachtelte Bedingungen durch die Verwendung von logischen Operatoren oder elif-Anweisungen (siehe n√§chsten Abschnitt) vereinfacht werden.

Beispiel f√ºr Vereinfachung einer verschachtelten Bedingung:

### Logische Operatoren in Bedingungen

Logische Operatoren erm√∂glichen es, mehrere Bedingungen zu kombinieren und komplexere Ausdr√ºcke zu erstellen.

#### üìù Die drei logischen Operatoren

1. **and**: Wahr, wenn beide Operanden wahr sind
2. **or**: Wahr, wenn mindestens ein Operand wahr ist
3. **not**: Kehrt den Wahrheitswert des Operanden um

#### üìù Wahrheitstabellen

**and-Operator**:

| A | B | A and B |
| --- | --- | --- |
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

**or-Operator**:

| A | B | A or B |
| --- | --- | --- |
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | False |

**not-Operator**:

| A | not A |
| --- | --- |
| True | False |
| False | True |

#### üìù Beispiele f√ºr logische Operatoren in Bedingungen

#### üìù Kurzschlussauswertung (Short-circuit Evaluation)

Python verwendet Kurzschlussauswertung f√ºr logische Operatoren:

- Bei `and`: Wenn der erste Operand `False` ist, wird der zweite Operand nicht ausgewertet, da das Ergebnis immer `False` sein wird.
- Bei `or`: Wenn der erste Operand `True` ist, wird der zweite Operand nicht ausgewertet, da das Ergebnis immer `True` sein wird.

````python
# Beispiel f√ºr and-Kurzschluss
x = 5
if x > 10 and x / 0 > 0:  # Der zweite Teil wird nicht ausgewertet, keine Division durch Null
    print("Bedingung erf√ºllt")

# Beispiel f√ºr or-Kurzschluss
y = 15
if y > 10 or y / 0 > 0:  # Der zweite Teil wird nicht ausgewertet, keine Division durch Null
    print("Bedingung erf√ºllt")
````

### üìù Beispiel 6: Debugging einer bedingten Anweisung


````python
def berechne_versandkosten(bestellwert, land):
    """
    Berechnet die Versandkosten basierend auf dem Bestellwert und dem Land.

    Bedingungen:
    - Bestellwert >= 50 ‚Ç¨ -> kostenloser Versand innerhalb Deutschlands
    - Bestellwert >= 100 ‚Ç¨ -> kostenloser Versand innerhalb der EU
    - Deutschland: 4,99 ‚Ç¨
    - EU: 9,99 ‚Ç¨
    - Restliche Welt: 19,99 ‚Ç¨
    """
    # Debug-Ausgaben zur Fehlerbehebung
    print(f"Debug: Bestellwert = {bestellwert}, Land = {land}")

    # Validierung der Eingaben
    if not isinstance(bestellwert, (int, float)) or bestellwert < 0:
        return "Fehler: Bestellwert muss eine positive Zahl sein."

    if not isinstance(land, str) or not land:
        return "Fehler: Land muss angegeben werden."

    land = land.lower()

    # Entscheidungslogik
    if land == "deutschland":
        if bestellwert >= 50:
            return 0.0  # Kostenloser Versand
        else:
            return 4.99
    elif land in ["frankreich", "italien", "spanien", "niederlande", "belgien", "√∂sterreich"]:
        if bestellwert >= 100:
            return 0.0  # Kostenloser Versand
        else:
            return 9.99
    else:
        return 19.99

# Testf√§lle zur √úberpr√ºfung
test_cases = [
    (30, "Deutschland"),
    (60, "Deutschland"),
    (80, "Frankreich"),
    (120, "Italien"),
    (50, "USA"),
    (-10, "Deutschland"),
    (100, "")
]

for bestellwert, land in test_cases:
    kosten = berechne_versandkosten(bestellwert, land)
    print(f"Bestellwert: {bestellwert} ‚Ç¨, Land: {land}, Versandkosten: {kosten} ‚Ç¨")
````
