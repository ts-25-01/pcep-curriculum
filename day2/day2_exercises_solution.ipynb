### Aufgabe 1: Variablen und Datentypen

Erstelle ein Programm, das folgende Variablen definiert und deren Werte und Typen ausgibt:
- Eine Ganzzahl
- Eine Fließkommazahl
- Einen String
- Einen booleschen Wert
- Verschiedene arithmetische Operationen mit diesen Variablen

```python
# Beispiellösung Aufgabe 1
ganzzahl = 42
kommazahl = 3.14
text = "Python"
ist_wahr = True

# Ausgabe der Werte und Typen
print(f"ganzzahl = {ganzzahl}, Typ: {type(ganzzahl)}")
print(f"kommazahl = {kommazahl}, Typ: {type(kommazahl)}")
print(f"text = {text}, Typ: {type(text)}")
print(f"ist_wahr = {ist_wahr}, Typ: {type(ist_wahr)}")

# Arithmetische Operationen
print(f"\nArithmetische Operationen:")
print(f"{ganzzahl} + {kommazahl} = {ganzzahl + kommazahl}")
print(f"{ganzzahl} * {kommazahl} = {ganzzahl * kommazahl}")
print(f"{text} * {ganzzahl // 10} = {text * (ganzzahl // 10)}")
```

### Aufgabe 2: String-Manipulation

Schreibe ein Programm, das einen String manipuliert:
1. Erstelle einen String mit deinem vollständigen Namen
2. Extrahiere den Vornamen und den Nachnamen mit Slicing
3. Zähle, wie viele Buchstaben dein Name insgesamt hat
4. Ersetze einen Buchstaben in deinem Namen durch einen anderen
5. Gib deinen Namen in Großbuchstaben aus

```python
# Beispiellösung Aufgabe 2
name = "Max Mustermann"

# Extrahieren von Vor- und Nachname
vorname = name[:3]  # "Max"
nachname = name[4:]  # "Mustermann"

# Zählen der Buchstaben
anzahl_buchstaben = len(name.replace(" ", ""))  # 12

# Ersetzen eines Buchstabens
neuer_name = name.replace("M", "N")  # "Nax Nustermann"

# Großbuchstaben
name_gross = name.upper()  # "MAX MUSTERMANN"

# Ausgabe
print(f"Vollständiger Name: {name}")
print(f"Vorname: {vorname}")
print(f"Nachname: {nachname}")
print(f"Anzahl Buchstaben: {anzahl_buchstaben}")
print(f"Name mit ersetztem Buchstaben: {neuer_name}")
print(f"Name in Großbuchstaben: {name_gross}")
```

### Aufgabe 3: Typkonvertierung

Schreibe ein Programm, das:
1. Zwei Zahlen vom Benutzer als Strings einliest
2. Diese in numerische Typen konvertiert
3. Verschiedene arithmetische Operationen durchführt
4. Die Ergebnisse als formatierte Strings ausgibt

```python

zahl1_str = input("Geben Sie eine ganze Zahl ein: ")
zahl2_str = input("Geben Sie eine Kommazahl ein: ")

# Konvertierung zu numerischen Typen
zahl1 = int(zahl1_str)
zahl2 = float(zahl2_str)

# Arithmetische Operationen
summe = zahl1 + zahl2
differenz = zahl1 - zahl2
produkt = zahl1 * zahl2
quotient = zahl1 / zahl2

# Ausgabe
print(f"Zahl 1: {zahl1} (Typ: {type(zahl1)})")
print(f"Zahl 2: {zahl2} (Typ: {type(zahl2)})")
print(f"Summe: {summe}")
print(f"Differenz: {differenz}")
print(f"Produkt: {produkt}")
print(f"Quotient: {quotient:.2f}")  # Formatierung auf 2 Nachkommastellen

# Konvertiere die Ergebnisse zurück zu Strings
ergebnis_str = f"Die Summe von {zahl1} und {zahl2} ist {summe}."
print(ergebnis_str)
```

### Aufgabe 4: Operatorpräzedenz

Berechne den Wert der folgenden Ausdrücke und erkläre die Auswertungsreihenfolge:
1. `2 + 3 * 4`
2. `(2 + 3) * 4`
3. `10 - 2 ** 3 + 5`
4. `20 / 4 / 2`
5. `8 + 4 // 3 - 2 * 2`

```python
# Beispiellösung Aufgabe 4
ausdruck1 = 2 + 3 * 4
ausdruck2 = (2 + 3) * 4
ausdruck3 = 10 - 2 ** 3 + 5
ausdruck4 = 20 / 4 / 2
ausdruck5 = 8 + 4 // 3 - 2 * 2

print(f"2 + 3 * 4 = {ausdruck1}")
print("Erklärung: Zuerst 3 * 4 = 12, dann 2 + 12 = 14")

print(f"\n(2 + 3) * 4 = {ausdruck2}")
print("Erklärung: Klammern zuerst: 2 + 3 = 5, dann 5 * 4 = 20")

print(f"\n10 - 2 ** 3 + 5 = {ausdruck3}")
print("Erklärung: Potenz zuerst: 2 ** 3 = 8, dann 10 - 8 + 5 = 7")

print(f"\n20 / 4 / 2 = {ausdruck4}")
print("Erklärung: Von links nach rechts: 20 / 4 = 5, dann 5 / 2 = 2.5")

print(f"\n8 + 4 // 3 - 2 * 2 = {ausdruck5}")
print("Erklärung: Erst 4 // 3 = 1 und 2 * 2 = 4, dann 8 + 1 - 4 = 5")
```

### Aufgabe 5: Logische Operatoren

Bestimme den Wert der folgenden logischen Ausdrücke:
1. `True and False`
2. `not (True and False)`
3. `True or (False and True)`
4. `(5 > 3) and (10 <= 10)`
5. `(5 == 5) or not (7 >= 9)`
6. `"Python" in "Python ist eine Programmiersprache" and 3 < 2`

```python
# Beispiellösung Aufgabe 5
ausdruck1 = True and False
ausdruck2 = not (True and False)
ausdruck3 = True or (False and True)
ausdruck4 = (5 > 3) and (10 <= 10)
ausdruck5 = (5 == 5) or not (7 >= 9)
ausdruck6 = "Python" in "Python ist eine Programmiersprache" and 3 < 2

print(f"True and False = {ausdruck1}")
print("Erklärung: False, da beide Operanden True sein müssen")

print(f"\nnot (True and False) = {ausdruck2}")
print("Erklärung: True, da (True and False) = False, und not False = True")

print(f"\nTrue or (False and True) = {ausdruck3}")
print("Erklärung: True, da der erste Operand bereits True ist (Kurzschlussauswertung)")

print(f"\n(5 > 3) and (10 <= 10) = {ausdruck4}")
print("Erklärung: True, da beide Vergleiche True sind")

print(f"\n(5 == 5) or not (7 >= 9) = {ausdruck5}")
print("Erklärung: True, da (5 == 5) bereits True ist")

print(f"\n\"Python\" in \"Python ist eine Programmiersprache\" and 3 < 2 = {ausdruck6}")
print("Erklärung: False, da 3 < 2 False ist")
```

### Aufgabe 6: Praxisaufgabe - Temperaturrechner

Erstelle einen Temperaturrechner, der:
1. Eine Temperatur in Celsius als String einliest
2. Diese in Fahrenheit umrechnet (Formel: F = C * 9/5 + 32)
3. Das Ergebnis formatiert ausgibt
4. Zusätzlich anzeigt, ob es sich um eine Temperatur über dem Gefrierpunkt handelt

```python
# Beispiellösung Aufgabe 6
# In einem echten Programm würdest du input() verwenden
celsius_str = input("Bitte gib die Temperatur in Celsius an")

# Konvertierung und Berechnung
celsius = float(celsius_str)
fahrenheit = celsius * 9/5 + 32

# Überprüfung des Gefrierpunkts
ueber_gefrierpunkt = celsius > 0  # z.B. bei 10 Grad Celsius, ist das 10 > 0, also wahr, bei -10 ist -10 > 0 falsch
print(ueber_gefrierpunkt) # Wenn über 0, dann ist dieser Wert True, d.h. wir printen True aus

# Ausgabe
print(f"{celsius}°C entspricht {fahrenheit:.2f}°F")

if ueber_gefrierpunkt:
    print("Diese Temperatur liegt über dem Gefrierpunkt (0°C).")
else:
    print("Diese Temperatur liegt am oder unter dem Gefrierpunkt (0°C).")

# Formatierter Output mit F-Strings
temperatur_info = f"""
Temperaturinformationen:
------------------------
Eingabe: {celsius}°C
Umgerechnet: {fahrenheit:.2f}°F
Status: {'Über' if ueber_gefrierpunkt else 'Am oder unter'} dem Gefrierpunkt
"""

print(temperatur_info)
```

### Aufgabe 7: Datentypen erkennen

Bestimme den Datentyp und den Wert der folgenden Ausdrücke:
1. `5 + 3.0`
2. `"5" + "3"`
3. `int("5") + int("3")`
4. `str(5) + str(3)`
5. `bool("False")`
6. `float("5") / 2`
7. `10 / 3`
8. `10 // 3`

```python
# Beispiellösung Aufgabe 7
ausdruck1 = 5 + 3.0
ausdruck2 = "5" + "3"
ausdruck3 = int("5") + int("3")
ausdruck4 = str(5) + str(3)
ausdruck5 = bool("False")
ausdruck6 = float("5") / 2
ausdruck7 = 10 / 3
ausdruck8 = 10 // 3

print(f"5 + 3.0 = {ausdruck1}, Typ: {type(ausdruck1)}")
print("Erklärung: 8.0, Typ: float (int + float ergibt float)")

print(f"\n\"5\" + \"3\" = {ausdruck2}, Typ: {type(ausdruck2)}")
print("Erklärung: '53', Typ: str (String-Verkettung)")

print(f"\nint(\"5\") + int(\"3\") = {ausdruck3}, Typ: {type(ausdruck3)}")
print("Erklärung: 8, Typ: int (Konvertierung zu int, dann Addition)")

print(f"\nstr(5) + str(3) = {ausdruck4}, Typ: {type(ausdruck4)}")
print("Erklärung: '53', Typ: str (Konvertierung zu str, dann Verkettung)")

print(f"\nbool(\"False\") = {ausdruck5}, Typ: {type(ausdruck5)}")
print("Erklärung: True, Typ: bool (nicht-leerer String ist True)")

print(f"\nfloat(\"5\") / 2 = {ausdruck6}, Typ: {type(ausdruck6)}")
print("Erklärung: 2.5, Typ: float (Division ergibt float)")

print(f"\n10 / 3 = {ausdruck7}, Typ: {type(ausdruck7)}")
print("Erklärung: 3.3333..., Typ: float (Division ergibt float)")

print(f"\n10 // 3 = {ausdruck8}, Typ: {type(ausdruck8)}")
print("Erklärung: 3, Typ: int (Ganzzahldivision)")
```
